## ANN

Creating an ANN Model(mimics of Human Brain)

*Neurons are used and connected to each other

Neuron Network has layers(Input,Hidden,Output)all are connected to each other internally.

*Fully Neural Network--All Previous nodes are connected to next nodes(layer)

*Every node should be associated with Activation function(used as gate)

*In Input Layer and Hidden Layer RELU activation fn is used.(Relu is applied on W*X and it keeps +ve value as it is and - value convert to 0)

*If Problem satement contains Binary classifiaction Output Layer me Sigmoid activation fn used and SOFTMAX will be used in O/p Layer when Multiclass Classification.

****it is not considered to use Neural Network everytime as it is Computationally Expensive
---It is used when environment is not static and changing ,for 3D objects,Speech,handwritten Recognizition.
--In this solution time dependent
---Ann Suffers Curse of Dimensionality when it comes to high resolution imges.

```python
import tensorflow as tf
import keras
```

```python
from keras.models import Sequential  #Layers are conected sequentially(FeedForward)
from keras.layers import Dense  #dense layer(Normal layer) is imported (Every Node of Privious layer connected to nodes of Next layers)
```

```python
deep_ann=Sequential()
```

```python
#Adding the Layers
deep_ann.add(Dense(units=6,kernel_initializer="uniform",activation='relu'))  #units=no.of nodes *Basically it is an Hiddeen layer
deep_ann.add(Dense(units=4,kernel_initializer="uniform",activation='relu'))
deep_ann.add(Dense(units=6,kernel_initializer="uniform",activation='relu'))
deep_ann.add(Dense(units=4,kernel_initializer="uniform",activation='relu'))  #kernel Intializer is stating that inially weight is uniform for all nodes
deep_ann.add(Dense(units=1,activation='sigmoid'))
#5 layers are added
```

```python
#Compiling the Model
deep_ann.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import pandas as pd
data=pd.read_csv("/content/Churn_Modelling.csv")
```

```python
data.describe()
```

```python
data.duplicated().sum()
```

```python
data.drop(["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1:]
```

```python
Y.head()
```

```python
data["Gender"].unique()
```

```python
X=pd.get_dummies(X,dtype=int,drop_first=True)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X_train ,X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)  #random state is used so that data dosent split randomly again(and same splitting of data points occur) and hence its Accuracy sustained.
```

```python
X_train.head()
```

```python
X_train.shape #Use 11 as input dim
```

```python
Y_train.head()
```

```python
#1.Standarization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train) #used for calculating Mean and S.D(fit_transform)
X_test=sc.transform(X_test) #transform just put saved mean and SD of X train.
```

```python
from keras.models import Sequential  #Layers are conected sequentially(FeedForward)
from keras.layers import Dense  #dense layer(Normal layer) is imported (Every Node of Privious layer connected to nodes of Next layers)
#Adding the Layers
deep_ann.add(Dense(units=6,kernel_initializer="uniform",activation='relu',input_dim=11))  #units=no.of nodes *Basically it is an Hiddeen layer(as units increses accuracy increases)
deep_ann.add(Dense(units=4,kernel_initializer="uniform",activation='relu'))
deep_ann.add(Dense(units=6,kernel_initializer="uniform",activation='relu'))
deep_ann.add(Dense(units=4,kernel_initializer="uniform",activation='relu'))  #kernel Intializer is stating that inially weight is uniform for all nodes
deep_ann.add(Dense(units=1,activation='sigmoid'))
#5 layers are added
deep_ann.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Model Training
deep_ann.fit(X_train,Y_train,batch_size=15,epochs=100) #batch size divide datset in batches *Epoch value train the dataset value times
```

```python
#Prediction
Y_pred=deep_ann.predict(X_test)
```

```python
Y_pred=(Y_pred>0.5)
```

```python
#Evaluate the performance
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```
