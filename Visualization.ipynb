# Visualization-->

```python
#Line Plot
import matplotlib.pyplot as plt
X = [1,2,3,4,6,7]
Y = [1,4,8,15,34,40]
plt.plot(X,Y,color="red")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Plot")
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt
X = [1,2,3,4,6,7]
Y = [1,4,8,15,34,40]
plt.scatter(X,Y,color="red")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot")
plt.show()
```

```python
#Bar plot --Vertical
sub=["Python","Os","DBMS","AI","CN","SE"]
count =[78,45,39,60,95,82]
plt.bar(sub,count,color ="red")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.title("Bar Plot")
plt.show()
```

```python
#Bar plot --Horizontal
sub=["Python","Os","DBMS","AI","CN","SE"]
count =[78,45,39,60,95,82]
plt.barh(sub,count,color ="red")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.title("Bar Plot")
plt.show()
```

```python
#Pie Plot
sub=["Python","Os","DBMS","AI","CN","SE"]
count =[78,45,39,60,95,82]
plt.pie(count,labels=sub,autopct="%0.2f%%")
plt.show()
```
