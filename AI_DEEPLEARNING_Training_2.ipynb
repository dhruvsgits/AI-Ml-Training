### KNN for Classification
It is used when an Problem Target coloum contain Discrete/Categorical data values(t/f yes/no cat/dog) Ex--Heart Disease,spam detection,image classification

```python
#Binary class,Multi class Classifier
#Models--Knn,SVm,DT
```

```python
#KNN Model-->
#Algo steps--1.Select the value of k
#2.Find the distance of new data point with the existing data
#3.Find the nearest distance
#4.
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

```

```python
data=pd.read_csv("/content/heart - Copy.csv")
data.head()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isnull().sum()
```

```python
data["trestbps"].fillna(data["trestbps"].mean(),inplace=True)
data["thalach"].fillna(data["thalach"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
#Dataset Splitting
import sklearn
from sklearn.model_selection import train_test_split
X_train ,X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)  #random state is used so that data dosent split randomly again(and same splitting of data points occur) and hence its Accuracy sustained.
```

```python
#1.Standarization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train) #used for calculating Mean and S.D(fit_transform)
X_test=sc.transform(X_test) #transform just put saved mean and SD of X train.
```

```python
#Model Creation
from sklearn.neighbors import KNeighborsClassifier
knnc=KNeighborsClassifier(n_neighbors=5)
knnc.fit(X_train,Y_train)
```

```python
#Model testing/Prediction
Y_pred=knnc.predict(X_test)
```

```python
#Evaluate the performance
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```
