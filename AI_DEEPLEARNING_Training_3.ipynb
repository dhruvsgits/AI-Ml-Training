```python
#Create intelligent model for the Prediction of Types of plant
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv("/content/iris - Copy.csv")
data.head()
```

```python
data.shape
data.describe()
data.info()
data.columns
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
data["sepal_width"].fillna(data["sepal_width"].mean(),inplace=True)
data["petal_length"].fillna(data["petal_length"].mean(),inplace=True)
data["petal_width"].fillna(data["petal_width"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
# Instead of using mean, use Time Series imputation to fill missing values in the 'species' column.
# It will do Forward Filling.
data["species"].fillna(method='ffill', inplace=True)
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1:]
```

```python
Y.head()
```

```python
#As it is Y we do Labell encoding
Y["species"].unique()
```

```python
Y.replace('setosa',0,inplace=True)
Y.replace('versicolor',1,inplace=True)
Y.replace('virginica',2,inplace=True)
```

```python
Y.head()
```

```python
#Dataset Splitting
import sklearn
from sklearn.model_selection import train_test_split
X_train ,X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)  #random state is used so that data dosent split randomly again(and same splitting of data points occur) and hence its Accuracy sustained.
```

```python
X_train.head() #same scale so no need for Feature Scaling
```

```python
#Model Training and Hyperparameteer Tunning
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
knnc=KNeighborsClassifier()
grid_param={"p":[1,2],"n_neighbors":[2,3,4,5,6,7,8]}
gm=GridSearchCV(knnc,grid_param)
gm.fit(X_train,Y_train)
print("Best Parameters for model",gm.best_params_)

```

```python
#Model Creation
knnc=KNeighborsClassifier(n_neighbors=4,p=2)
knnc.fit(X_train,Y_train)
```

```python
#Model testing/Prediction
Y_pred=knnc.predict(X_test)
```

```python
#Evaluate the performance
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```

```python
#take User Input and Predict category
sepal_length=float(input("Enter Sepal Length-->"))
sepal_width=float(input("Enter Sepal Width-->"))
petal_length=float(input("Enter Petal Length-->"))
petal_width=float(input("Enter Petal Width-->"))

user_input=[[sepal_length,sepal_width,petal_length,petal_width]]
predicted_Plant=knnc.predict(user_input)
print("Prediction is -->",predicted_Plant)
```
