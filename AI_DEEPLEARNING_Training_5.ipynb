```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
deep_cnn=Sequential()

deep_cnn.add(Conv2D(32,(3,3),input_shape=(64,64,3),activation='relu'))
deep_cnn.add(MaxPooling2D(pool_size=(2,2))) #used to reduce image metrices in 2*2(default)metrices

deep_cnn.add(Conv2D(32,(3,3),activation='relu'))
deep_cnn.add(MaxPooling2D(pool_size=(2,2)))

deep_cnn.add(Flatten())

deep_cnn.add(Dense(units=32,activation='relu'))
deep_cnn.add(Dense(units=32,activation='relu'))

deep_cnn.add(Dense(units=1,activation='sigmoid'))

```

```python
deep_cnn.summary()
```

```python
deep_cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```
